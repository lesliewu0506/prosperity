```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_paths = [
    "data/1.csv",
    "data/2.csv",
    "data/3.csv",
    "data/4.csv",
    "data/5.csv",
    "data/6.csv",
]
dfs = []
for i, file_path in enumerate(file_paths):
    df_day = pd.read_csv(file_path, sep=";")
    df_day["timestamp"] = df_day["timestamp"] + i * 999900
    df_day["day"] = i
    dfs.append(df_day)
df = pd.concat(dfs, ignore_index=True)
df.tail()
```

```python
pepper = df[df["product"] == "INTARIAN_PEPPER_ROOT"]
pepper.head()
```

```python
osmium = df[df["product"] == "ASH_COATED_OSMIUM"]

osmium.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharex=True)

axes[0].scatter(pepper["timestamp"], pepper["mid_price"], s =5)
axes[0].set_title("INTARIAN_PEPPER_ROOT")
axes[0].set_xlabel("timestamp")
axes[0].set_ylabel("mid_price")
# axes[0].set_ylim(12500, 14500)

axes[1].scatter(osmium["timestamp"], osmium["mid_price"], s=5)
axes[1].set_title("ASH_COATED_OSMIUM")
axes[1].set_xlabel("timestamp")
axes[1].set_ylabel("mid_price")
# axes[1].set_ylim(9950, 10050)
plt.tight_layout()
plt.show()
```
